# MIT PROx Data Science Case Study 1.2 - Identifying Faces with PCA

```python
# Import required libraries
from PIL import Image
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
import os
from glob import glob
from sklearn.preprocessing import normalize
import plotly
import plotly.express as px
```

```python
# Load images
image_dir = './instructors'
n_instructors = 14

def get_image(image_path, width, height, mode):
    
    image = Image.open(image_path)

    return np.array(image.convert(mode))


def get_batch(image_files, width, height, mode):
    data_batch = np.array(
        [get_image(sample_file, width, height, mode) for sample_file in image_files]).astype(np.float32)
    return data_batch

instructor_images = get_batch(glob(os.path.join(image_dir, "*.jpg"))[:n_instructors], 300, 300, "L")
```

```python
# Visualize some images
from matplotlib import pyplot as plt
%matplotlib inline

fig = plt.figure(figsize = (10, 10))
for i in range(n_instructors):
    ax = fig.add_subplot(3, 5, i+1, xticks = [], yticks = [])
    ax.imshow(instructor_images[i], cmap = "gray")
```

```python
# Example
plt.imshow(instructor_images[0], cmap = "gray")
```

## Normalizing the data and applying PCA

```python
# Appending the images into a vector

image_matrix = []
for i in range(n_instructors):
    image = instructor_images[i].reshape(300 * 300,)
    image_matrix.append(image)
image_matrix = np.asarray(image_matrix)

normal_image = normalize(image_matrix)
```

```python
# Mean face
mean_face = np.mean(normal_image, axis = 0)
plt.imshow(mean_face.reshape(300, 300), cmap = "gray")
```

```python
# The differences of each image off the mean face
difference_face = normal_image - mean_face.reshape(1, -1)

fig = plt.figure(figsize = (10, 10))
for i in range(n_instructors):
    ax = fig.add_subplot(3, 5, i+1, xticks = [], yticks = [])
    ax.imshow(difference_face[i].reshape(300, 300), cmap = "gray")
```

## Manual Calculation of the Eigenfaces

```python
# The instruction of manually calculating the eigenface can be found here: https://sandipanweb.wordpress.com/2018/01/06/eigenfaces-and-a-simple-face-detector-with-pca-svd-in-python/

# Covariance Matrix C, considering N^2 * N^2 would be too large, use M * M as the dimensions of the covariance matrix instead
C = np.matmul(difference_face, difference_face.T)

# Eigenvectors of C
eigen_C = np.linalg.eig(C)
eigen_value = eigen_C[0]
eigen_vector_C = eigen_C[1]

# Non-normalized eigenvectors calculated from C
eigen_v = np.matmul(difference_face.T, eigen_vector_C)
eigen_norm = normalize(eigen_v)
```

```python
plt.imshow(eigen_norm.T[0].reshape(300, 300), cmap = 'gray')
```

```python
# Applying PCA
model = PCA(n_components = n_instructors)
model.fit(normal_image)
pca_image = model.transform(normal_image)
```

```python
# The eigenfaces
#Example Normalized with eigenfaces
fig = plt.figure(figsize = (10, 10))
for i in range(n_instructors):
    ax = fig.add_subplot(3, 5, i+1, xticks = [], yticks = [])
    ax.imshow(model.components_[i].reshape(300, 300), cmap = "gray")
```

## Reconstruction faces

```python
# Projected face


components = model.transform(normal_image)
projected = model.inverse_transform(components)
```

```python
fig, axes = plt.subplots(10,1,figsize=(20,80))
for i, ax in enumerate(axes.flat):
    ax.imshow(projected[i].reshape(300,300),cmap="gray")
```
